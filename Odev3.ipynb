```python
import sqlite3

conn = sqlite3.connect("ornek_veritabani.db")
cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS ogrenciler (
                  id INTEGER PRIMARY KEY,
                  ad TEXT,
                  notlar INTEGER);''')

ogrenci_verileri = [
    ("Barış", 100),
    ("Mehmet", 78),
    ("Zeynep", 95),
    ("Ahmet", 88),
    ("Elif", 76),
    ("Selin", 23),
    ("Hakan", 32),
    ("Ebru", 22),
    ("Kemal", 11),
    ("Sevgi", 47),
    ("Tolga", 82),
    ("Ceren", 45),
    ("Murat", 55),
    ("Gül", 55),
    ("Berk", 30),
    ("Emre", 81),
    ("Dilara", 74),
    ("Kadir", 17),
    ("Gizem", 77),
    ("Can", 2),
    ("Aslı", 0),
    ("Sude", 72),
    ("Onur", 0),
    ("Duygu", 71),
]

for ogrenci in ogrenci_verileri:
    cursor.execute("INSERT INTO ogrenciler (ad, notlar) VALUES (?, ?)", ogrenci)
conn.commit()
conn.close()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
conn = sqlite3.connect("ornek_veritabani.db")
df = pd.read_sql_query("SELECT * FROM ogrenciler", conn)
print(df)
df.plot(x="ad", y="notlar", kind="bar")
plt.title("Öğrenci Notları")
plt.xlabel("Öğrenci Adı")
plt.ylabel("Notlar")
plt.show()
conn.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

conn = sqlite3.connect("ornek_veritabani.db")
df = pd.read_sql_query("SELECT * FROM ogrenciler", conn)
notlar_grup = df['notlar'].value_counts()

plt.pie(notlar_grup, labels=notlar_grup.index, autopct='%1.1f%%')
plt.title("Öğrenci Notlarının Dağılımı")
plt.axis('equal')
plt.show()
conn.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

conn = sqlite3.connect("ornek_veritabani.db")
df = pd.read_sql_query("SELECT * FROM ogrenciler", conn)
plt.hist(df['notlar'], bins=10, edgecolor='k')
plt.title("Öğrenci Notlarının Dağılımı")
plt.xlabel("Notlar")
plt.ylabel("Frekans")
plt.show()
conn.close()
```
